```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
diab = load_diabetes()
diab
```

```python
df = pd.DataFrame(diab.data, columns=diab.feature_names)
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(x=df['age'],y=df['sex'],hue=df['bmi'])
```

```python
sns.pairplot(df, hue='sex', size=3)
plt.show()
```

```python
plt.hist(df['age'],bins=20, rwidth=0.8, color='red')
plt.show()
```

```python
sns.boxplot(df['age'], orient='h', color='red')
plt.show()
```
